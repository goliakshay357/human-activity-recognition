```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climbing = pd.read_csv("climb/LIS3DH Accelerometer.csv")
climbing.head()
```

```python
jumping=pd.read_csv("jump/LIS3DH Accelerometer.csv")
jumping.head()
```

```python
running = pd.read_csv("run/LIS3DH Accelerometer.csv")
running.head()
```

```python
walking = pd.read_csv("walk/LIS3DH Accelerometer.csv")
walking.head()
a = np.asarray(walking)
a
```

```python
columns = ["time","x","y","z"]
```

```python
jumping_df = np.asarray(jumping)
jumping_df = pd.DataFrame(data= jumping_df , columns = columns)
jumping_df
```

```python
climbing_df = np.asarray(climbing)
climbing_df = pd.DataFrame(data= climbing_df , columns = columns)
climbing_df
```

```python
walking_df = pd.DataFrame(data = a , columns = columns) 
walking_df
```

```python
running_df = np.asarray(running)
running_df = pd.DataFrame(data= running_df , columns = columns)
running_df
```
