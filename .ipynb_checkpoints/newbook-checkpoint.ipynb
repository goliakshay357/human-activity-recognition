```python
import os
import glob
# climbing = pd.read_csv("data/climb/temp/climb1.csv")
X = ["climb","jump","jog","walk"]

for i in range(len(X)):
    def concatenate(indir="/home/goli/Desktop/human_a_Det/data/" + X[i]+ "/temp",outfile = "/home/goli/Desktop/human_a_Det/output/"+X[i]+".csv"):
        os.chdir(indir)
        fileList = glob.glob("*.csv")
        dfList = []
        colnames = ["time","x","y","z"]
        for filename in fileList:
            print(filename)
            df = pd.read_csv(filename,header = None)
            df.drop(df.head(50).index, inplace=True)
            df.drop(df.tail(50).index, inplace=True)
            dfList.append(df)
        concatDf = pd.concat(dfList,axis = 0)
        concatDf.columns = colnames
        concatDf.to_csv(outfile,index = None)
    concatenate()
```

```python
final_output = pd.read_csv("/home/goli/Desktop/human_a_Det/output/"+"final_sol"+".csv")
final_output

x = final_output.iloc[:,:-1].values
y = final_output.iloc[:,9].values

from sklearn.preprocessing import LabelEncoder
label_encoder_y = LabelEncoder()
y = label_encoder_y.fit_transform(y)

print(len(final_output))
```

```python
final_output.corr()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(x, y, test_size = 0.2, random_state = 1,shuffle = True)
```

```python
#MODAL
from sklearn import svm
from sklearn import metrics
types=['linear']
for i in types:
    model=svm.SVC(kernel=i)
    model.fit(train_X,train_Y)
    prediction=model.predict(test_X)
    print('The accuracy of the SVM kernel ',i,'is',metrics.accuracy_score(prediction,test_Y))
accuracy_SVM = metrics.accuracy_score(prediction,test_Y)
```

```python
#modal
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=4)
model.fit(train_X,train_Y)
prediction=model.predict(test_X)
print('The accuracy of the SVM kernel ',i,'is',metrics.accuracy_score(prediction,test_Y))
accuracy_SVM = metrics.accuracy_score(prediction,test_Y)
```

```python
#modal using sklearn Logistic regressions
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model = LogisticRegression()
model.fit(train_X,train_Y)
prediction=model.predict(test_X)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_Y))
accuracy_Logistic = metrics.accuracy_score(prediction,test_Y)
```
