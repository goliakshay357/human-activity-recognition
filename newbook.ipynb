```python
import os
import glob
# climbing = pd.read_csv("data/climb/temp/climb1.csv")
X = ["climb","jump","jog","walk"]

for i in range(len(X)):
    def concatenate(indir="/home/goli/Desktop/human_a_Det/data/" + X[i]+ "/temp",outfile = "/home/goli/Desktop/human_a_Det/output/"+X[i]+".csv"):
        os.chdir(indir)
        fileList = glob.glob("*.csv")
        dfList = []
        colnames = ["time","x","y","z"]
        for filename in fileList:
            print(filename)
            df = pd.read_csv(filename,header = None)
            df.drop(df.head(50).index, inplace=True)
            df.drop(df.tail(50).index, inplace=True)
            dfList.append(df)
        concatDf = pd.concat(dfList,axis = 0)
        concatDf.columns = colnames
        concatDf.to_csv(outfile,index = None)
    concatenate()
```
